```python
import numpy as np
```

```python
def load_embedding(path):
    embedding_index = {}
    with open(path, encoding='utf-8') as f:
        for index, line in enumerate(f):
            if index == 0:
               continue
            values = line.split(' ')
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embedding_index[word] = coefs
            print(word, coefs)
    f.close()
    return embedding_index
```
